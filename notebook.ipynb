```python
import pandas as pd

df = pd.read_csv('/content/StudentsPerformance.csv')

# عرض أول 5 صفوف
df.head()
```

```python
# شكل الأعمدة وأنواع البيانات
df.info()

# إحصائيات عامة
df.describe()

# هل في بيانات ناقصة؟
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# توزيع درجات الرياضيات
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
plt.show()

```

```python
# Boxplot حسب النوع
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.show()

# Boxplot حسب الكورس التحضيري
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Scores by Test Preparation Course')
plt.show()

# Boxplot حسب مستوى تعليم ولي الأمر
plt.figure(figsize=(10,5))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.title('Math Scores by Parental Education Level')
plt.xticks(rotation=45)
plt.show()



```

```python
df.groupby('gender')['math score'].mean()
df.groupby('test preparation course')['math score'].mean()
df.groupby('parental level of education')['math score'].mean().sort_values(ascending=False)
```
